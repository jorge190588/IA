```python
import tensorflow as TensorFlow #Importacion de la Libreria de tensorflow y se le asigna un Alias de nombre TensorFlow.
import matplotlib.image as MatplotlibImagen #Importacion de la Libreria matplotlib.image y se le asigna un Alias de nombre MatplotlibImagen.
import numpy as Numpy #Importación de la libreria propia de Python numpy y se le asigna un Alias de nombre Numpy.
import matplotlib.pyplot as MatplotlibPyplot #Importacion de la Libreria matplotlib.pyplot y se le asigna un Alias de nombre MatplotlibPyplot.
import matplotlib.pyplot as plt
TamanoDeLaImagen = 224 #Tamaño de la imagen DE 224 Pixeles.
def RedimensionarImagen(RutaDeDireccionDeArchivo): #Declaramos la variable de nombre RedimensionarLaImagen y le pasamos una ruta de dirección como parametro.
    Imagen = [] #Array que retornara la imagen redimensionada.
    TensorFlow.reset_default_graph() #Resetea el gráfico global que estabá predeterminado.
    ImagenOriginal = TensorFlow.placeholder(TensorFlow.float32, (None, None, None)) #Se Declara un tensor con Nombre de ImagenOriginal que almacenará la imagen de entra de tipo float.
    ImagenDelTensor = TensorFlow.image.resize_images(ImagenOriginal, (TamanoDeLaImagen, TamanoDeLaImagen), TensorFlow.image.ResizeMethod.NEAREST_NEIGHBOR) #Se Cambia el tamaño de la imagen redimensionada.         
    with TensorFlow.Session() as IniciarSesion: #Se inicia la sesion de TensorFlow y se le asigna un Alias con nombre IniciarSesion.
        IniciarSesion.run(TensorFlow.global_variables_initializer()) #Se inicia las variables globales de la Sesion.
#Cada imagen se cambia de tamaño de forma individual ya que diferentes imágenes pueden ser de diferentes tamaños.

        for Contador, RutaDeLaImagen in enumerate(RutaDeDireccionDeArchivo): #Ciclo para recorrer la Ruta de Direccion de Archivo con un Contador Automatico.
            print("Imagen No: "+str(Contador+1)+" de: "+str(len(RutaDeDireccionDeArchivo))+" Guardada en: "+str(RutaDeLaImagen))
            #Imprime el Numero de Imagen, la Cantidad que hay redimensionadas, y la ruta de direccion donde esta guardada la imagen.
            ImagenDeEntrada = MatplotlibImagen.imread(RutaDeLaImagen)[:, :, :3]   #La Variable ImagenDeEntrada Carga el valor de la Ruta de la imagen      
            ImagenConNuevoTamano = IniciarSesion.run(ImagenDelTensor, feed_dict = {ImagenOriginal: ImagenDeEntrada}) #La Variable ImagenConNuevoTamaño Obtiene el nuevo tamaño de la imagen
            print('Tamaño de la Imagen De Entrada: {0[0]} X {0[1]} Pixeles'.format(ImagenDeEntrada.shape)) #Imprime el tamaño nuevo de la imagen de entrada.
            ImagenReparada= Numpy.multiply(ImagenConNuevoTamano, 1.0 / 255.0) #Repara la Imagen con la Formula de Matriz de la libreria Numpy y la guarda en la variable de ImagenReparada
            print('Tamaño de la Imagen Reparada: {0[0]} X {0[1]} Pixeles'.format(ImagenReparada.shape)) #Imprime el tamaño reparado de la imagen
            Imagen.append(ImagenReparada) # Arrega al array de ListaDeImagen la imagen que se a ridemensionado.
    Imagen = Numpy.array(Imagen, dtype = Numpy.float32) # Convierte la ListaDeImagen a tipo Numpy flotante
    return Imagen #Retorna los datos almacenados en el array de ListaDeImagen

ImagenObtenida = RedimensionarImagen(['/notebooks/imagenes/ropa/blusa1.jpg']) #Se obtiene la imagen de la ruta establecida y se guarda en la variable ImagenObtenida
```

```python
import cv2 # libreria de complementos de python de  OpenCV Version 2
def AgregarRuidoGaussiano(ImagenObtenida): #Se Define la funcion de Ruido Gaussiano y se recibe como parametros la Imagen Obtenida del proceso anterior.
    ImagenConRuidoGaussiano = [] #Se Declara un Array para retonar el proceso de Ruido Gaussiano.    
    Fila, Columna, _ = ImagenObtenida[0].shape #Se declara la matriz de datos para la Imagen Obtenida.
    
    for TecnicaDeLaImagen in ImagenObtenida: #Ciclo que recorre la ImagenObtenida para realizar la tecnica de Condicion de Iluminacion.
       
        Tecnica = Numpy.random.random((Fila, Columna, 1)).astype(Numpy.float32) #se asigna valores aletaroios a la variable Tecnica.
        Tecnica = Numpy.concatenate((Tecnica, Tecnica, Tecnica), axis = 2) #Concatena los valores asiganos en el eje 2.    
        ImagenConGaussiano = cv2.addWeighted(TecnicaDeLaImagen, 0.25, 0.75 * Tecnica, 0.25, 2) #realiza la tecnica de ilumiacion con el proceso de ruido gausiano
        
        ImagenConRuidoGaussiano.append(ImagenConGaussiano) #agregar la tecnica de ruido gaussiano a la Imagen.
    ImagenConRuidoGaussiano = Numpy.array(ImagenConRuidoGaussiano, dtype = Numpy.float32) #Convierte la Imagen a tipo Numpy flotante.
    return ImagenConRuidoGaussiano #retorna el resultado de proceso de ruido gausiano
  
ImagenConRuidoGaussiano = AgregarRuidoGaussiano(ImagenObtenida) #Agrega el Proceso de la tecnica a la imagen redimensianada que se tenia en memoria.
```

```python
# Mostrar imagenes

for imagen in ImagenObtenida: #Ciclo que recorre la variable ImagenObtenida que tiene asignada la imagen redimensionada.  
    imgplot = MatplotlibPyplot.imshow(imagen) #Renderiza la imagen de tipo Numpy que esta guardada en la variable
    print("Imagen Redimensionada Obtenida") #Imprime mensasje de encabezado
    MatplotlibPyplot.show() #Muestra la imagen    
    i=0
for imagen in ImagenConRuidoGaussiano: #Ciclo que recorre la variable ImagenConRuidoGaussiano que tiene es la que guardo el proceso de la tecnica. 
    i=i+1
    imgplot = MatplotlibPyplot.imshow(imagen) #Renderizar la imagen de tipo Numpy que esta guardada en la variable
    print("Imagen Con Ruido Gaussiano") #Imprime mensasje de encabezado
    MatplotlibPyplot.savefig('DESTINO/Iluminacion/MyNewImage'+str(i)+'.png')
    MatplotlibPyplot.show() #Muestra la imagen
    
```
