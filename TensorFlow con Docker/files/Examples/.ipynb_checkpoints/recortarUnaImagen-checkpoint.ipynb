```python
import tensorflow as tf
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import os
import numpy as np


di='/notebooks/fperas'
# First, load the image again
dir_path = os.listdir(di)
print(dir_path)


for pera in dir_path:
    print(pera)  
    raw_image_data = mpimg.imread(di+'/'+pera,"")
    pera = tf.placeholder("float", [None, None, None])
    #slice(entrada,inicio,tamao,name=None)
    r = tf.slice(pera,[0,0,0],[-1,-1,1])
    g = tf.slice(pera,[0,0,1],[-1,-1,1])
    b = tf.slice(pera,[0,0,2],[-1,-1,1])


    #r = tf.scalar_mul(0.2126,r)
    #g = tf.scalar_mul(0.7152,g)
    #b = tf.scalar_mul(0.0722,b)
    grey = tf.add(r,tf.add(g,b))

    #concat(valores,eje,name='concat{})
    #out = tf.concat(2, [r,g,b])
    #out = tf.concat(2, [grey,grey, grey])
    
    out = tf.concat([r,g,b],2)
    out = tf.cast(out, tf.uint8)
    
    image = tf.placeholder("uint8", [None, None, None])

    # tf.slice(image2, [inicio, inicio, inicio], [filas, -1, -1])

    slice = tf.slice(image, [80, 0, 0], [120, -1, -1])
    
    with tf.Session() as session:
        result = session.run(slice, feed_dict={image: raw_image_data})
        print(result.shape)

    plt.imshow(result)
    plt.show()
    
     
    with tf.Session() as session:
        result = session.run(out, feed_dict={pera: raw_image_data})
        plt.imshow(result)
        plt.show()
        
    
    
```
