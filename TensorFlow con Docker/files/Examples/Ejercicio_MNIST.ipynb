```python
# importando el dataset
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
%matplotlib inline

from tensorflow.examples.tutorials.mnist import input_data
mnist = input_data.read_data_sets("MNIST_data/", one_hot=True)
```

```python
train_dataset = mnist.train.images
train_labels = mnist.train.labels

test_dataset = mnist.test.images
test_labels = mnist.test.labels

validation_dataset = mnist.validation.images
validation_labels = mnist.validation.labels

indice = 13
sample = train_dataset[indice]
sample = sample.reshape(28,28)

# Obtener la posición del número en el array y convertirlo a int
#array_labels = train_labels[indice]
#print("Etiqueta -> ",array_labels)
#posicion = np.where(array_labels == 1)
#print(int(posicion[0]))
#print(train_labels[1])

numero_buscar = 5
veces_encontrado = 0

for i in range(0,100):        
    array_labels = train_labels[i]
    posicion = np.where(array_labels == 1)
    int_posicion = int(posicion[0])
    
    #print("posicion",int_posicion,i)
    if(int_posicion == numero_buscar):                
        #print("Entro")
        veces_encontrado = veces_encontrado + 1
        plt.imshow(np.reshape(train_dataset[i],(28,28)), cmap='Greys_r')
        plt.show()
    
    
print("Número buscado ->",numero_buscar," Veces encontrado ->",veces_encontrado)    
```
